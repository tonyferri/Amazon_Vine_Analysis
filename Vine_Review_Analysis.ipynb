```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load
data_to_load = "Resources/vine_table.csv"

# Read the data files and store them in a pandas DataFrame.
vine_data_df = pd.read_csv(data_to_load)
vine_data_df
```

```python
vine_data_df.dtypes
```

```python
# total_votes count is equal to or greater than 20
twenty_total_votes_df = vine_data_df[(vine_data_df["total_votes"] >= 20)]
twenty_total_votes_df
```

```python
# number of helpful_votes divided by total_votes is equal to or greater than 50%.
twenty_total_votes_df["vote_%"] = (twenty_total_votes_df["helpful_votes"] / twenty_total_votes_df["total_votes"]) * 100
# twenty_total_votes_df

fifty_percent_helpful_df = twenty_total_votes_df[(twenty_total_votes_df["vote_%"] >= 50)]
fifty_percent_helpful_df
```

```python
# retrieve all rows where a review was written as part of the Vine program (paid), vine == 'Y'.
vine_paid_df = fifty_percent_helpful_df[(fifty_percent_helpful_df["vine"] == "Y")]
vine_paid_df
```

```python
# retrieve all rows where a review was written NOT part of the Vine program (paid), vine == 'N'.
vine_unpaid_df = fifty_percent_helpful_df[(fifty_percent_helpful_df["vine"] == "N")]
vine_unpaid_df
```

```python
vine_paid_df.count()
```

```python
# retrieve rows where vine is paid ("Y") and a five star review is given
vine_paid_5star_df = vine_paid_df[(vine_paid_df["star_rating"] >= 5)]
vine_paid_5star_df.count()
```

```python
# retrieve rows where vine is paid ("Y") and a four star review is given
vine_paid_4star_df = vine_paid_df[(vine_paid_df["star_rating"] >= 4)]
vine_paid_4star_df.count()
```
